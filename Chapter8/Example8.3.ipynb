Example 8.3: Analysis of the Simple Harmonic Oscillator.

Vibrations of a mass hanging from a linear spring are governed by the linear differential equation: 

$m \ddot{x} + kx = 0$, where $m$ is the mass, $k$ is the spring constant, and $x$ is the displacement of the mass from equilibrium. 

Give a phase plane analysis of this simple harmonic oscillator. 

As you already know very well, it is easy to solve this equation. But that's exactly what makes linear equations so special. For the nonlinear equations that we are ultimately interested in, it's impossible to find analytical solutions and we will need to revert to the methods we have been discussing thus far in this chapter to qualitatively understand them before analyzing them numerically. 

The motion in the phase plane is determined by a vector field that comes from the differential equation. To find the vector field, construct the two equations:

$\dot{x} = v$, i.e. the velocity, and:

$\dot{v} = -\frac{k}{m} x$, i.e. the rate of change of the velocity, as we did in the previous chapter. 

We can also write them as:

$\dot{x} = v$,

$\dot{v} = - \omega^2 x$, where $\omega^2 = k/m$, the square of the angular frequency. 

This system assigns a vector $(\dot{x}, \dot{v}) = (v, - \omega^2 x)$ *at each point* $(x,v)$, and therefore represents a *vector field* on the phase plane. 

```python
import matplotlib.pyplot as plt # import matplotlib, a conventional module name is plt
import numpy as np

# Meshgrid 
x, y = np.meshgrid(np.linspace(-2, 2, 10),  
                   np.linspace(-2, 2, 10)) 

# Directional vectors 
u = y
v = -0.4*x 

fig, ax = plt.subplots() # create the elements required for matplotlib. This creates a figure containing a single axes.

# set the labels and titles:
ax.set_xlabel(r'$x$', fontsize=20) # set the x label
ax.set_ylabel(r'$v$', fontsize=20) # set the y label
ax.set_title('Vector field for SHO', fontsize=10) # set the title 
# aspect:
ax.set_aspect(0.5)
# Creating plot 
plt.quiver(x, y, u, v, color='red', units='xy', scale=1, zorder=10, width=0.035) 
ax.plot(0,0, marker='o', color='k')

ax.set_xticks([])
ax.set_yticks([])

plt.show() # show the plot here
```

Just as we did in one dimension, it is helpful to visualize the vector field in terms of an imaginary fluid. Imagine that the fluid is flowing steadily on the phase plane with local velocity $(\dot{x}, \dot{v}) = (v, - \omega^2 x)$. 

Then to find the trajectory starting at $(x_0, v_0)$, we place an imaginary particle at $(x_0, v_0)$ and watch how it is carried around by the flow. 

The origin is special: like the eye of a hurricane, the origin is a fixed point. But a phase point starting anywhere else would circulate around the origin and eventually return to its starting point. Such trajectories form closed orbits. 

```python
import matplotlib.pyplot as plt # import matplotlib, a conventional module name is plt
import numpy as np
from matplotlib.markers import MarkerStyle
from matplotlib.patches import Ellipse
from matplotlib.transforms import Affine2D

fig, ax = plt.subplots() # create the elements required for matplotlib. This creates a figure containing a single axes.
# set the labels and titles:
ax.set_xlabel(r'$x$', fontsize=20) # set the x label
ax.set_ylabel(r'$v$', fontsize=20) # set the y label
ax.set_title('SHO Phase Portrait', fontsize=10) # set the title 
# aspect:
ax.set_aspect(1)
# Creating plot 
ellipse = Ellipse(
    xy=(0, 0),
    width=100,
    height=40,
    angle=0,
    facecolor="none",
    edgecolor="r"
)
ax.add_patch(ellipse)
# Plot an arrow marker at the end point of minor axis
vertices = ellipse.get_co_vertices()
t = Affine2D().rotate_deg(ellipse.angle)
ax.plot(
    vertices[0][0],
    vertices[0][1],
    color="r",
    marker=MarkerStyle(">", "full", t),
    markersize=10
)

# Creating plot 
ellipse = Ellipse(
    xy=(0, 0),
    width=50,
    height=20,
    angle=0,
    facecolor="none",
    edgecolor="r"
)
ax.add_patch(ellipse)
# Plot an arrow marker at the end point of minor axis
vertices = ellipse.get_co_vertices()
t = Affine2D().rotate_deg(ellipse.angle)
ax.plot(
    vertices[0][0],
    vertices[0][1],
    color="r",
    marker=MarkerStyle(">", "full", t),
    markersize=10
)
ax.set_xticks([])
ax.set_yticks([])

ax.plot(0,0, marker='o', color='k')


plt.show() # show the plot here 
```

This is known as the phase portrait of the system. 

The fixed point $(x,v)=(0,0)$ corresponds to static equilibrium of the system: the mass is at rest at its equilibrium position and will remain there forever. The closed orbits correspond to period motions, i.e. oscillations of the mass. 

The shape of the closed orbits has an interesting physical interpretation: they are ellipses given by the equation $\omega^2 x + v^2 = C$, where $C\geq 0$ is a constant. This is equivalent to the conservation of energy!
