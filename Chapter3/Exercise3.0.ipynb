# Exercise 3.0: Random numbers according to an exponential function

Use uniform random numbers in the interval $[0,1]$ to generate random numbers according to the exponential function $f(t) = \exp{(-t)} / (1-\exp(-1))$ via the inversion method. 

Compare to the actual function in the interval $t \in [0,1]$.


Solution: The CDF is given by: 

$F(t) = \int^t_0 \mathrm{d} t \exp{(-t)} = (1-\exp{(-t)})/(1-\exp(-1)) = (1-\exp{(-t)})/\alpha$

Then, we need to choose random numbers according to $t = \ln (1- \alpha y) $

```python
import random 
import math
import matplotlib.pyplot as plt # import matplotlib
import numpy as np

t = []

alpha = 1 - math.exp(-1)
# We have t = ln(y) and we pick y in [0,1] 
for i in range(10000): # generate 1000 points
    y = random.random() # generate random numbers in [0,1]
    t.append( -math.log(1-alpha * y) )


# create the real distribution for comparison. Use numpy"
tcomp = np.linspace(0,1,1000)
ycomp = np.exp(-tcomp)/alpha
    
# create a histogram of x:
fig, ax = plt.subplots() # create the elements required for matplotlib. This creates a figure containing a single axes.

# set the labels and titles:
ax.set_xlabel(r'$t$', fontsize=20) # set the x label
ax.set_ylabel(r'$P(t)$', fontsize=20) # set the y label. Note that the 'r' is necessary to remove the need for double slashes. You can use LaTeX! 
ax.set_title('Histogram', fontsize=20) # set the title 

# make a histogram of the values, density=True normalizes the histogram
ax.hist(t, label='generated values via inversion method', density=True) 

# compare to how we expect it to look like:
ax.plot(tcomp, ycomp, label='expected shape of $f(t) = \exp(-t) $')

# construct the legend:
ax.legend(loc='upper center')  # Add a legend

plt.show() # show the plot here
```
